```python
import instaloader

# Get instance
L = instaloader.Instaloader()


L.interactive_login("OWNUSERNAME")      # (ask password on terminal)
```

```python
profile = instaloader.Profile.from_username(L.context, "TARGET_USERNAME")

print("{} is following these profiles:".format(profile.username))
followeeCount =  0
for followee in profile.get_followees():
    print(followee.username)
    followeeCount += 1
    
print("\nTotal following: {}".format(followeeCount))
```
